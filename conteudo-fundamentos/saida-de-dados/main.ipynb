## MÃ©todo format()

```python
faturamento = 2000
print("O faturamento da loja foi R${}".format(faturamento) + ".")
```

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo

print("O faturamento da loja foi R${}. O custo da loja foi R${}. Assim, o lucro da loja foi R${}".format(faturamento, custo, lucro) + ".")
```

## F-strings -> Recomendado

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo

print(f"O faturamento da loja foi R${faturamento}. O custo da loja foi R${custo}. Assim, o lucro da loja foi R${lucro}.")
```
