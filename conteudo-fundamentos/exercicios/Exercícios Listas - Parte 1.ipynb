## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]

vendas_anual = vendas_1sem + vendas_2sem

maior_valor_vendas_ano = max(vendas_anual)
menor_valor_vendas_ano = min(vendas_anual)

print(f"Maior valor de vendas: {maior_valor_vendas_ano}\nMenor valor de vendas: {menor_valor_vendas_ano}")

# Unir listas → +
# Encontrar um indice a partir do seu valor na lista → index()
```

## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
vendas_anual = vendas_1sem + vendas_2sem

maior_valor_vendas_ano = max(vendas_anual)
menor_valor_vendas_ano = min(vendas_anual)

mes_melhor_venda = vendas_anual.index(maior_valor_vendas_ano)
mes_pior_venda = vendas_anual.index(menor_valor_vendas_ano)

print(f"O melhor mês do ano foi: {meses[mes_melhor_venda]} com {maior_valor_vendas_ano} vendas\nO pior mês do ano foi: {meses[mes_pior_venda]} com {menor_valor_vendas_ano} vendas")

faturamento_anual = sum(vendas_anual)
porcentagem_do_faturamento_anual_maior_valor = (maior_valor_vendas_ano / faturamento_anual) * 100
print(f"O faturamento anual foi de {faturamento_anual}\nPorcentagem do maior valor de vendas anual em relação ao faturamento anual: {porcentagem_do_faturamento_anual_maior_valor:.2f}%")
```

## 3. Crie uma lista com o top 3 valores de vendas do ano (sem fazer "no olho")

Dica: o método remove retira um item da lista.

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]

vendas_anual = vendas_1sem + vendas_2sem
print(vendas_anual)

top_3_vendas = sorted(vendas_anual, reverse=True)[:3]
print(f"TOP 3 VENDAS: {top_3_vendas}")

# sorted(vendas) → ordena a lista
# reverse=True → ordem decrescente (do maior para o menor)
# [:3] → pega os 3 primeiros elementos
```
